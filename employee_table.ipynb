```python
%sql
CREATE OR REPLACE TABLE skillbrew_lab.default.employees(
  employee_id INT,
  name String,
  department String,
  salary DOUBLE,
  hire_date DATE
)
```

```python
%sql
-- Now let’s make a change and look at the history: give  every engineering employee a 11% raise
UPDATE skillbrew_lab.default.employees
SET salary = salary *1.10
WHERE department = "Engineering";
```

```python
%sql
OPTIMIZE skillbrew_lab.default.employees;
-- This merges small Parquet files into larger ones. Faster reads.

VACUUM skillbrew_lab.default.employees;
-- This cleans up files older than 7 days.
-- After VACUUM, you can't time-travel to versions whose files were removed.
```

```python
%sql
-- Standard view: runs the query every time
CREATE OR REPLACE VIEW skillbrew_lab.default.high_earners AS
SELECT 
  name,
  department,
  salary
FROM 
  skillbrew_lab.default.employees
WHERE salary > 90000;
```

```python
%sql
-- Temp view: gone when you close the notebook
CREATE OR REPLACE TEMP VIEW session_stats AS 
SELECT
  department,
  COUNT(*) AS headcount, AVG(salary) AS avg_sal
FROM skillbrew_lab.default.employees
GROUP BY department;

-- Materialized view (syntax to know; needs paid tier or DLT)
-- CREATE MATERIALIZED VIEW analytics.daily_revenue AS
-- SELECT 
--   order_date,
--   SUM(amount) AS revenue FROM orders GROUP BY order_date;
```

```python
%sql
-- Create well documented product table
CREATE OR REPLACE TABLE skillbrew_lab.default.products(
  product_id INT COMMENT 'Unique identifier for each product',
  product_name STRING COMMENT 'WHat we call this product in catalogs',
  category STRING COMMENT 'Product line (electronics, Home, etc)',
  price DOUBLE COMMENT 'Current retail price in USD',
  is_active BOOLEAN COMMENT 'False if dicontinued'
)
COMMENT 'Product master list maintained by merchandising team' ;
```
