```python
%sql
CREATE OR REPLACE TABLE skillbrew_lab.default.employees(
  employee_id INT,
  name String,
  department String,
  salary DOUBLE,
  hire_date DATE
)
```

```python
%sql
INSERT INTO skillbrew_lab.default.employees VALUES
(1,"Alice Johnson", "Engineering", 95000, "2022-01-15"),
(2,"Bob Smith", "Marketing", 72000, "2022-03-20"),
(3,"Carol WIlliams", "Engenieering", 105000, "2021-06-10"),
(4,"David Brown", "Sales", 68000, "2023-01-05"),
(5,"Eve Davis", "Engenieering", 98000, "2022-09-12")
```

```python
%sql
-- Now let’s make a change and look at the history: give  every engineering employee a 11% raise
UPDATE skillbrew_lab.default.employees
SET salary = salary *1.10
WHERE department = "Engineering";
```

```python
%sql
--Check what happened behind the scenes
DESCRIBE HISTORY skillbrew_lab.default.employees;
```

```python
%sql 
-- Go back in time and see the original salaries
SELECT * FROM skillbrew_lab.default.employees VERSION AS OF 3;
```

```python
%sql
-- Compare with what we have now
SELECT * FROM skillbrew_lab.default.employees VERSION AS OF 5
```

```python
%sql
OPTIMIZE skillbrew_lab.default.employees;
-- This merges small Parquet files into larger ones. Faster reads.

VACUUM skillbrew_lab.default.employees;
-- This cleans up files older than 7 days.
-- After VACUUM, you can't time-travel to versions whose files were removed.
```
